```python
import matplotlib.pyplot as plt
```

```python
from scipy.stats import norm
```

```python
P=[2,4,4,4,5,5,7,9]
```

```python
plt.hist(P)
plt.title("histogram")
std = np.std(P)
mean = np.mean(P)
plt.plot(norm.pdf(P,mean,std))
plt.show()
```

```python
x_axis = np.arange(-10,10,0.001)
```

```python
plt.plot(x_axis,norm.pdf(x_axis, 0.2))
plt.grid()
plt.title("kurva")
plt.show()
```

```python
mu=10.0
sigma=2.0
```

```python
data = np.random.randn(100000)*sigma+mu
```

```python
hx, hy, _ =plt.hist(data, bins=50, color="lightblue")
plt.ylim(0.0,max(hx)+0.05)
plt.title("Generate random number \n from a norman distribution with python")
plt.grid()
plt.show()
```
